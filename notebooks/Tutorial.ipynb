# Fast-SL Community Screener: E. coli & Salmonella Tutorial
Welcome to the Fast-SL Community pipeline. This tutorial will walk you through running an exhaustive synthetic lethality screen on two bacterial models sharing a single environment. 

By utilizing Parsimonious FBA (pFBA) to mathematically prune silent backup pathways, this pipeline evaluates hundreds of thousands of genetic combinations in minutes.


```python
# Import the custom fastsl engine and data analysis libraries
import fastsl
import pandas as pd
import matplotlib.pyplot as plt

# Optional: Make plots look nice
plt.style.use('ggplot')
```

### 1. Setting Up the Models
First, we point the pipeline to the BiGG XML models for *Escherichia coli* and *Salmonella enterica*. 
*(Note: Because this notebook is inside the `notebooks/` folder, we use `../models/` to look up one level).*


```python
model_ecoli = "../models/iML1515.xml"
model_salm = "../models/STM_v1_0.xml"
```

### 2. Running the Exhaustive Screen
Now we launch the pipeline. We will set `cores=14` for parallel processing and establish a `death_fraction` of `0.01`. This means if a knockout combination drops a bacterium's growth rate below 1% of its healthy Wild-Type growth, it is classified as dead.

```python
# Run the core engine
results_df = fastsl.run_pipeline(
    model1=model_ecoli, name1="Ecoli",
    model2=model_salm, name2="Salm",
    cores=14,
    death_fraction=0.01,
    output_csv="../Ecoli_Salm_Tutorial.csv"
)

# Display the first 5 lethal combinations found
results_df.head()
```

### 3. Visualizing the Lethality
Once the Double Lethals are calculated, we can immediately visualize how the knockouts affected the community. Did they cause total ecological collapse, or did they only starve one specific species?

```python
# Count the occurrences of each lethality type
lethality_counts = results_df['Lethality_Type'].value_counts()

# Create a bar chart
plt.figure(figsize=(8, 5))
lethality_counts.plot(kind='bar', color=['#e63946', '#457b9d', '#1d3557'])

# Add labels and title
plt.title('Distribution of Synthetic Lethality (E. coli + Salmonella)')
plt.ylabel('Number of Double Lethals')
plt.xticks(rotation=0)

# Display the plot
plt.show()
```
